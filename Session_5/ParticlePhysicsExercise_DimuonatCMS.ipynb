# Particle Physics Exercise - Dimuon spectra at the LHC

## Introduction

The [Large Hadron Collider](https://home.cern/science/accelerators/large-hadron-collider) (LHC) at CERN, located in the French-Swiss border, start operations in 2011 and it was built to perform precisions tests of the Standard Model of particle physics and to search for new physics. The LHC is a hadron collider of 27 km of radius, accelerating beams of hadrons, protons or heavy ions, in opposite directions close to speed of light. 
These two beams cross each other in 4 major interaction points where particle detectors are built. Two major multipurpose detectors: ATLAS and CMS, and two more specialized ones: LHCb for b-physics and ALICE for heavy ions. The LHC is located in the same tunnel of the previous LEP experiment. The two proton beams are accelerated in a set of steps, starting from a bottle of hidrogen where they are accelerated in the linear collider LINAC2, continuing with the BOOSTER, PS, SPS and finally to the LHC. The CERN accelerator complex is illustrated in the figure below.

The [Compact Muon Solenoid](https://cms.cern/) (CMS) is a multipurpouse detector, and it is located in the French town of Cessy. It is the second largest detector in the LHC ring and, as most of the modern particle detectors, it was designed to have different subdetectors to identify different particles coming from the collisions. It's solenoid creates a magnetic field strong enough that each layer of the detector was built one right after the other, in a compact manner. A cartoon of the different components of the CMS detector is shown on the right of the figure below.

CMSdetector.png

<div align="center">Figure 2: Diagram of the CMS detector.</div>

Stable particles created from the proton-proton collisions interact with the different components of the CMS detector. A simple schema of how these particles interact with the detector is shown in Fig. 3. 

Muons are charged leptons which will pass through the entire detector (blue curvy line in Fig.3). These charge particles leave signals (tracks) mainly in two components of the detector: the tracker (most inner component) and the muon chamber (most outside component), both containing different layers. Different algorithms are used to reconstruct of these trajectories. One of these algorithms creates so-called *Global Muons* where muons are reconstucted starting from tracks within the muon chamber and going inside to match tracks in the tracker. A different approach creates so-called *Tracker Muons*, where the reconstruction starts from tracks in the tracker and continues with the tracks in the muon chamber. More information in Ref. [2].

<div align="center">Figure 3: Schema of how particles interact with the CMS detector.</div>

The reconstruction of muons is cleaner than other stable particles. In that sense, it is simpler to look at the dimuon mass spectra in order to identify heavier unstable particles decaying into a pair of muons. Figure 4 shows this dimuon spectrum with several resonances, each corresponding to a different particle. This spectrum allows to rediscover several particles from the SM and also to calibrate several of the components in our detector.

## Description of work to do

1. To get more familiar with how these events look like inside the CMS detector, several event displays can be seen [here](http://opendata.cern.ch/record/303). Play with this tool and get to know how this physics topology looks like.

1. Then, look at the data samples. These files contain preselected events with mostly two or more muons. Compute the invariant mass of the two muons in the files and try to reproduce the dimuon spectra.

1. Now, let's concentrate in the Z boson. MonteCarlo simulations of this process is also provided. Here, your first task is to determine which trigger (or set of triggers) are the best to use in data. The triggers are a set of filters that are used during the data taking to a priori select interesting events.

1. Once you learn which triggers are the best to your signal, study the MC simulations further to identify if any additional kinematic variables can be used to enhance your signal. Can your selection be more optimal? Look at the dimuon spectrum and see if the shape of your signal can be improved.

1. Apply these set of selections (triggers+kinematics) to the data and plot again the dimuon spectra in the Z boson region (Dimuon mass above 50 GeV). Then, we need to model the background and signal components with functional forms. Try to model the background (falling part of the specrta) with a exponential-like function and the signal (peak) with a gaussian or double crystal function. (If you have problems with the signal part, you can use the MC simulations properly scaled to the cross section and luminosity) 

1. After modelling the signal and the background, compute the cross section of the Z boson. How does this compared to the Z boson cross section from the [this measurement](http://cms-results.web.cern.ch/cms-results/public-results/publications/SMP-12-011/index.html)?

1. If you have more time, you can also compared the Z boson mass and width with the numbers in the paper.
